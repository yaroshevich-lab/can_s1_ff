```python
import os
from rdkit import Chem
```

```python
REF_FILE = 'ecn_ff/45D.pdb'
INPUT_FOLDER = 'scan_structures_utf8' # Где лежат кривые файлы
OUTPUT_FOLDER = 'scan_remapped'
```

```python
ref = Chem.MolFromPDBFile(REF_FILE, removeHs=False)
```

```python
for file in os.listdir(INPUT_FOLDER):
    if not file.endswith('.pdb'):
        continue
    probe = Chem.MolFromPDBFile(f'{INPUT_FOLDER}/{file}', removeHs=False)
    match = probe.GetSubstructMatch(ref)
    new_probe = Chem.RenumberAtoms(probe, list(match)) 
    for i in range(ref.GetNumAtoms()):
        res_info = ref.GetAtomWithIdx(i).GetPDBResidueInfo()
        new_probe.GetAtomWithIdx(i).SetMonomerInfo(res_info)
    output_path = os.path.join(OUTPUT_FOLDER, file)
    writer = Chem.PDBWriter(output_path)
    writer.write(new_probe)
    writer.close()
```
